```python
import time,random

def easy_sort(seq):
    start_time=time.time()
    for i in range(len(seq)-1):
        min=i
        for j in range (i+1,len(seq)):
            if seq[j]<seq[min]:
                min=j
        seq[i],seq[min]=seq[min],seq[i]
    end_time=time.time()
    print("所用时间为",end_time-start_time)
```

```python
def quick_sort(seq):
    p=seq[0]
    left=[]
    right=[]
    for i in range(1,len(seq)):
        if seq[i]<=p:
            left.append(seq[i])
        else: right.append(seq[i])
    if left:
        left=quick_sort(left)
    if right:
        right=quick_sort(right)
    return left+[p]+right
start_time2=time.time()
```

```python
import random
def count_numbers_dict(seq): #词典法统计
    numbers_dict = {}
    for number in seq:
        if number in numbers_dict:
            numbers_dict[number] += 1   #如果词典中有键为number，则该number的数量+1
        else:  numbers_dict[number] = 1   #否则，词典中加入键number，其数量为1  
    return numbers_dict

#测试用
number=[random.randint(0,100001) for i in range (10000)]
num=sorted(number)  #内置函数排序
table = count_numbers_dict(num)
for i, item in enumerate(table):
    print("数为:",item,"数量为:",table[item])  #item表示为数字，table[item]表示为该数字的数量！！
```
